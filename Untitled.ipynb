```python
candf=pd.read_csv("Canada.csv")
```

```python
candf
```

```python
candf.drop(['AREA','REG','DEV','Type','Coverage'],inplace=True,axis=1)
```

```python
candf
```

```python


candf.rename(columns = {"OdName": "Country", 
                     "AreaName":"Continent",
                       "RegName":"Region"}, 
                                 inplace = True) 
```

```python
candf
```

```python
candf=candf.set_index('Country')
```

```python
candf
```

```python
candf=candf.drop(['year','Total'],axis=1)
```

```python
candf['Total']=candf.sum(axis=1)
```

```python
candf
```

# plot a line graph of immigration from Haiti

```python
years=list(map(str,range(1980,2013)))
```

```python
haiti=candf.loc['Haiti',years]
haiti
```

```python
import matplotlib.pyplot as plt
haiti.plot(kind='line')

plt.title("Immigration from Haiti")
plt.xlabel('Years')

plt.ylabel('Immigrants')



plt.text(28,6000,"2010 Earthquake")

```

# Compare the immigrants from India and China from 1980 to 2013

```python
india=candf.loc['India',years]
china=candf.loc['China',years]
```

```python
india.plot(kind='bar',color='r')
china.plot(kind='bar',color='b')
plt.legend()
```
